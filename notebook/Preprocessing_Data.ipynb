### 3.2 Transection Amount Risk

```python
fig, axes = plt.subplots(1, 2, figsize = (14,5))
colors = ['#2ecc71', '#e74c3c']
# 1. Distribution plot: Seaborn handles the 'subsetting' and 'legend' for you
sns.histplot(data = train_df, x = 'amt', hue = 'is_fraud', bins = 100, palette = colors, element= 'step', fill= True, stat= 'density', common_norm= False, ax = axes[0])
axes[0].set_ylim(0, None)
axes[0].set_xlim(0, 1000)
axes[0].set_title('Transaction Amount Distribution', fontweight='bold', fontsize = 14)
axes[0].legend(title='Class', labels=['Fraud', 'Legitimate'])

#2. Box plot
sns.boxplot(data = train_df, x = 'is_fraud', y = 'amt', palette= colors, ax = axes[1])
axes[1].set_title('Amount by Class', fontweight = 'bold', fontsize = 14)
axes[1].set_ylim(0, None)
axes[1].set_xticklabels(['Legitimate', 'Fraud'])

plt.tight_layout()
plt.show()

# Stats Summary
print("--- Amount Statistics ---")
display(train_df.groupby('is_fraud')['amt'].describe().round(2))
```

- Based on the histplot, legitimate transactions have a significantly higher density than fraudulent ones, concentrated in the $0–$300 range, while fraudulent transactions can extend above $1,000.
- The mean fraudulent transaction amount ($531.32) is approximately 8x higher than legitimate transactions ($67.67).
- Fraudulent transactions range from $1.06 to $1,376, while legitimate transactions have a much wider range of $1 to $28,948. This suggests fraudsters avoid extremely large amounts to evade detection systems.

Conclusion: The amt feature provides good class separability between fraudulent and legitimate transactions, making it a valuable feature for fraud detection


### 3.3 Category Risk

```python
# 1. Aggration (Double check the rate is multiplied)
cat_fraud = train_df.groupby('category')['is_fraud'].agg(total = 'count', fraud_rate = lambda x: x.mean() * 100).sort_values('fraud_rate')
cat_fraud
```

```python
# 2. Create Plot
fig, axes = plt.subplots(1, 2, figsize = (14,5))

# Chart 1: Rate
cat_fraud['fraud_rate'].plot(kind = 'barh',ax = axes[0], color = 'red', edgecolor = 'black')
axes[0].set_title('Fraud Rate by Merchant Category (%)', fontweight='bold')
axes[0].set_xlabel('Percentage (%)')

# Chart 2: Volume
cat_fraud['total'].plot(kind = 'barh',ax = axes[1], color = 'blue', edgecolor = 'black')
axes[1].set_title('Total Transaction Volume', fontweight='bold')
axes[1].set_xlabel('Count')

plt.tight_layout()
plt.show()
```

- While online categories (shopping_net, misc_net) show the highest fraud rates (~1.5–1.75%), their transaction volumes are relatively low.
- High-volume categories like gas_transport have moderate fraud rates (~0.45%) but may contribute more fraud cases in absolute terms. Although category shows some signal — particularly online vs. offline channels — the differences are not large enough to be a strong standalone predictor of fraud.

### 3.4 Temporal Risk Pattern

#### 3.4.1 Overview of temporal Pattern of risk by hour, day, month

```python
df = train_df.copy()
df['hour'] = df['trans_date_trans_time'].dt.hour
df['day_name'] = df['trans_date_trans_time'].dt.day_name()
df['month'] = df['trans_date_trans_time'].dt.month
```

```python
# Ploting
fig, axes = plt.subplots(1, 3, figsize = (20,5))
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Chart 1: Hourly(bar)
sns.barplot(data = df, x = 'hour', y = 'is_fraud', ax = axes[0], color ='blue', errorbar=None)
axes[0].set_title('Fraud Rate by Hour', fontweight ='bold')

# Chart 2: Day of week (bar)
sns.barplot(data = df, x = 'day_name', y = 'is_fraud', order = days_of_week, ax = axes[1], color = 'red', errorbar= None)
axes[1].set_title('Fraud Rate by Day', fontweight ='bold')

# Chart 3: Monht (linechar)
sns.lineplot(data = df, x = 'month', y = 'is_fraud', ax = axes[2], color = 'green', marker = 'o', errorbar= None)
axes[2].set_xticks(range(1,13))
axes[2].set_title('Fraud Rate by Month', fontweight = 'bold')

for ax in axes:
    ax.set_yticklabels(['{:,.1%}'.format(x) for x in ax.get_yticks()])
    ax.set_ylabel('Fraud Rate')

plt.tight_layout()
plt.show()
```

`Plot 1 – Fraud Rate by Hour`
- Fraudulent transactions peak sharply between 10 PM and 3 AM, reaching rates ~40x higher than daytime hours. This pattern suggests perpetrators deliberately exploit low-monitoring windows when cardholders are asleep and real-time oversight is minimal.

`Plot 2 – Fraud Rate by Day`
- Fraud rate rises steadily from Monday, peaking on Friday (~0.70%), likely due to higher transaction volumes mid-to-late week providing better camouflage. The drop on weekends aligns with reduced consumer activity rather than increased detection.

`Plot 3 – Fraud Rate by Month`
- Fraud rate is highest in January–February (~0.80%), consistent with post-holiday/Lunar New Year spending surges that inflate transaction volumes and create cover for fraudulent activity. Rates decline and remain volatile through the rest of the year, with minor spikes possibly linked to promotional periods.

#### 3.4.2 Hight Risk Fraud Window

```python
# Prepare two metrix
# 1. Rate Metrix
rate_data = df.pivot_table(index = 'day_name', columns = 'hour', values = 'is_fraud', aggfunc= 'mean').reindex(days_of_week)

# 2. Volume Metrix
volume_data = df.pivot_table(index = 'day_name', columns = 'hour', values = 'is_fraud', aggfunc= 'count').reindex(days_of_week)

# Ploting Heatmap
plt.figure(figsize=(20,8))

sns.heatmap(rate_data, cmap = 'coolwarm', annot = volume_data, fmt = 'd', cbar_kws= {'label': 'Fraud_Rate'})
plt.title('Fraud Rate (Color) vs Total Transaction Volume (Numbers)', fontweight='bold', fontsize=16)
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')

plt.tight_layout()
plt.show()
```

`Plot 4 – Fraud Rate vs. Transaction Volume Heatmap (Hour × Day)`
- The heatmap reveals a critical risk concentration at hours 22–23 across all weekdays, particularly on Thursday and Friday, where fraud rates reach their peak (dark red cells, >0.03). 
- Notably, these high-fraud windows coincide with relatively low transaction volumes compared to daytime hours, meaning each individual transaction carries a disproportionately higher fraud risk. 
- The 0–3 AM window also shows elevated fraud rates despite low volume, reinforcing that off-hours are systematically exploited. Daytime hours (5–21h) display uniformly low fraud rates (blue), suggesting that transaction volume alone does not drive fraud.

>Timing and day-of-week interaction effects are the dominant risk factors.
